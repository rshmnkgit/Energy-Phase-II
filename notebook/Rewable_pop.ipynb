### Energy consumption collection created in mongo

```python
#Default port used by MongoDB is 27017
conn='mongodb://localhost:27017'
client=pymongo.MongoClient(conn)
#Decleare the database
db=client['renewable_energy']
```

```python
#Declare the collection
col_engcons=db['energyconsumption']

df=energy_cons.to_dict('records')
col_engcons.insert_many(df)
```

```python
for row in db.energyconsumption.find():
    pprint(row)
    print("------------------------")
```

```python
# Energy production collection created in mongo
#Declare the collection
col_engprod=db['energyproduction']

energy_prod=energy_prod.to_dict('records')
col_engprod.insert_many(energy_prod)
```

```python
for row in db.energyproduction.find():
    pprint(row)
    print("------------------------")
```

```python
# save the population data frame into the mongo cloud
#Declare the collection
col_population=db['Populationdata']

pop_data=pop_df.to_dict('records')
col_population.insert_many(pop_data)
```

```python
col_uspop = db["USPopulation"]
col_uspop.insert_many(us_pop.to_dict('records'))
```
